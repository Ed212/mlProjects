```python
def nxProb(txt,word):
    # Takes in a string text and a word Returns a dctionary whose keys are the words that come after 
    # word and whose values are the number if times those words appear after taht word appears 
    chars = '."?!'
    lst = txt.strip(chars).split(word,1)[1].split()
    #print(lst)
    dictt = {}
    for word in lst: 
     if  word in dictt.keys():
        dictt[word] += 1
     else:
        dictt[word] = 1
    return dictt   

nxProb(sample_memo,"into")
```
